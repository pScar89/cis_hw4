```python
avg_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
avg_scores.plot(kind='bar')
```

```python
sns.catplot(data=df, x='race/ethnicity', kind='count', height=5, aspect=2)
plt.show()
```

```python
plot = sns.catplot(data=df, x='race/ethnicity', kind='count', height=5, aspect=2)
plot.set_xticklabels(rotation=45)
plt.show()
```

```python
sns.scatterplot(data=df, x='writing score', y='reading score', hue='test preparation course')
plt.show()
```

```python
df.plot.scatter(x='writing score', y='reading score', c=df['test preparation course'].map({'completed': 'blue', 'none': 'orange'}))
plt.show()
```

```python
df.plot.scatter(x='writing score', y='reading score', c=df['test preparation course'].map({'completed': 'blue', 'none': 'orange'}), figsize=(10, 6))
plt.show()
```
